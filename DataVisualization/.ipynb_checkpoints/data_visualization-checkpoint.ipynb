### Problem 1

The file `anscombe.npy` contains Anscombe’s quartet.
For each section of the quartet,
- Plot the data as a scatter plot on the box [0, 20] × [0, 13].
- Use `scipy.stats.linregress()` to calculate the slope and intercept of the least squares regression line for the data and its correlation coefficient (the first three return values).
- Plot the least squares regression line over the scatter plot on the domain $x\in[0,20]$.
- Report the mean and variance in $x$ and $y$, the slope and intercept of the regression line, and the correlation coefficient. Compare these statistics to those of the other sections.
- Describe how the section is similar to the others and how it is different.

```python
#load the file and store each columns
A = np.load("anscombe.npy")
x1 = A[:,0]
y1 = A[:,1]
x2 = A[:,2]
y2 = A[:,3]
x3 = A[:,4]
y3 = A[:,5]
x4 = A[:,6]
y4 = A[:,7]

#slops ,interception, and correlation coefficients of each linear regression line
s1, i1, r1 = lrg(x1,y1)[0:3]
s2, i2, r2 = lrg(x2,y2)[0:3]
s3, i3, r3 = lrg(x3,y3)[0:3]
s4, i4, r4 = lrg(x4,y4)[0:3]
#means, variances
meanx1 = np.mean(x1)
meany1 = np.mean(y1)
meanx2 = np.mean(x2)
meany2 = np.mean(y2)
meanx3 = np.mean(x3)
meany3 = np.mean(y3)
meanx4 = np.mean(x4)
meany4 = np.mean(y4)

varx1 = np.var(x1)
vary1 = np.var(y1)
varx2 = np.var(x2)
vary2 = np.var(y2)
varx3 = np.var(x3)
vary3 = np.var(y3)
varx4 = np.var(x4)
vary4 = np.var(y4)


#graph each quartet
domain = np.linspace(0,20,100)
plt.subplots_adjust(hspace = 0.8, wspace = 0.4)
plt.subplot(221)
plt.scatter(x1,y1)
plt.plot(domain, s1*domain + i1, "g")
plt.title("quartet I")

plt.subplot(222)
plt.scatter(x2,y2)
plt.plot(domain, s2*domain + i2, "g")
plt.title("quartet II")

plt.subplot(223)
plt.scatter(x3,y3)
plt.plot(domain, s3*domain + i3, "g")
plt.title("quartet III")

plt.subplot(224)
plt.scatter(x4,y4)
plt.plot(domain, s4*domain + i4, "g")
plt.title("quartet IV")

plt.show()
#report the data
print("the means for colomn 1 is " + str(meanx1))
print("the means for colomn 2 is " + str(meany2))
print("the means for colomn 3 is " + str(meanx2))
print("the means for colomn 4 is " + str(meany2))
print("the means for colomn 5 is " + str(meanx3))
print("the means for colomn 6 is " + str(meany3))
print("the means for colomn 7 is " + str(meanx4))
print("the means for colomn 8 is " + str(meany4))
print("the variances for colomn 1 is " + str(varx1))
print("the variances for colomn 2 is " + str(vary2))
print("the variances for colomn 3 is " + str(varx2))
print("the variances for colomn 4 is " + str(vary2))
print("the variances for colomn 5 is " + str(varx3))
print("the variances for colomn 6 is " + str(vary3))
print("the variances for colomn 7 is " + str(varx4))
print("the variances for colomn 8 is " + str(vary4))
print("the slope for quartet I is " + str(s1))
print("the slope for quartet II is " + str(s2))
print("the slope for quartet III is " + str(s3))
print("the slope for quartet IV is " + str(s4))
print("the intercept for quartet I is " + str(i1))
print("the intercept for quartet II is " + str(i2))
print("the intercept for quartet III is " + str(i3))
print("the intercept for quartet IV is " + str(i4))
print("the correlation coefficient for quartet I is " + str(s1))
print("the correlation coefficient for quartet II is " + str(s2))
print("the correlation coefficient for quartet III is " + str(s3))
print("the correlation coefficient for quartet IV is " + str(s4))

print("Even the points are very different, all four graphs have similiar mean, variance, slope, intercept, and correlation coefficient")

```

### Problem 2
The $n + 1$ Bernstein basis polynomials of degree $n$ are defined as follows.

$$b_{n,v}(x) = \binom{n}{v}x^v(1-x)^{n-v},\quad v=0,\ 1,\ \ldots,\ n$$

Plot the first $10$ Bernstein basis polynomials ($n = 0,\ 1,\ 2,\ 3$) as small multiples on the domain $[0, 1] \times [0, 1]$.
Label the subplots for clarity, adjust tick marks and labels for simplicity, and set the window limits of
each plot to be the same.
Consider arranging the subplots so that the rows correspond with $n$ and the columns with $v$.

(Hint: The constant $\binom{n}{v} = \frac{n!}{v!(n-v)!}$ is called the binomial coefficient and can be efficiently computed with `scipy.special.binom()` or `scipy.misc.comb()`.)

```python
# Plot the first 10 Bernstein basis polynomials.
x = np.linspace(-1, 1, 200)

for n in range(4):
    for v in range(n+1):
        #define the polynomial and plot it
        y = lambda x: binom(n,v)*x**v*(1-x)**(n-v)
        plt.subplot(4,4,n*4+v+1)
        plt.subplots_adjust(hspace = 1.5, wspace = 0.4)
        plt.plot(x, y(x))
        plt.xlabel("v = " + str(v))
        plt.ylabel("n = " + str(n))
        plt.xlim(-0.1,1.1)
        plt.ylim(-0.1,1.1)
        plt.tight_layout()
        
plt.show()

```

### Problem 3

The file `MLB.npy` contains measurements from over 1,000 recent Major League Baseball players, compiled by UCLA.
Each row in the array represents a different player; the columns are the player’s height (in inches), weight (in pounds), and age (in years), in that order.

Create several visualizations to show the correlations between height, weight, and age in the MLB data set.
Use at least one scatter plot.
Adjust the marker size, plot a regression line, change the window limits, and use small multiples where appropriate.

```python
#load the data and save each column
height, weight, age = np.load("MLB.npy").T
plt.subplots_adjust(hspace = 1.5, wspace = 0.4)

#plot the correlation between weights and heights
plt.subplot(321)
domain1 = np.linspace(65,85,100)
s1, i1, r1 = lrg(height,weight)[0:3]
plt.plot(domain1, s1*domain1 + i1, "g")
plt.xlabel("height(inches)")
plt.ylabel("weight(pounds)")
plt.title("weights/heights")
plt.scatter(height, weight, s =1)

#plot the correlation between weights and ages
plt.subplot(322)
domain2 = np.linspace(20,50,100)
s2, i2, r2 = lrg(age,weight)[0:3]
plt.plot(domain2, s2*domain2 + i2, "g")
plt.xlabel("age(years)")
plt.ylabel("weight(pounds)")
plt.title("weights/age")
plt.scatter(age, weight, s =1)
#plot the correlation between heights and heights
plt.subplot(323)
s3, i3, r3 = lrg(age,height)[0:3]
plt.plot(domain2, s3*domain2 + i3, "g")
plt.xlabel("age(years)")
plt.ylabel("height(inches)")
plt.title("heights/age")
plt.scatter(age, height, s =1)

#plot the histgram for age
plt.subplot(324)
plt.hist(age)
plt.xlabel("age")
plt.ylabel("#age")
plt.title("#age")
#plot the histgram for heights
plt.subplot(325)
plt.hist(height, bins =50)
plt.xlabel("height")
plt.ylabel("#height")
plt.title("#height")
#plot the histgram for weights
plt.subplot(326)
plt.hist(weight, bins =50)
plt.xlabel("weight")
plt.ylabel("#weight")
plt.title("#weight")

plt.show()




```

### Problem 4

The file `earthquakes.npy` contains data from over 17,000 earthquakes between 2000 and 2010 that were at least a 5 on the Richter scale.
Each row in the array represents a different earthquake; the columns are the earthquake’s date (as a fraction of the year), magnitude (on the Richter scale), longitude, and latitude, in that order.

Describe the data with two or three visualizations, including line plots, scatter plots, and histograms as appropriate.
Your plots should answer the following questions:
1. How many earthquakes happened every year?
2. How often do stronger earthquakes happen compared to weaker ones?
3. Where do earthquakes happen? Where do the strongest earthquakes happen?

Hint: Use `plt.axis("equal")` or `ax.set_aspect("equal")` to fix the aspect ratio, which may improve comparisons between longitude and latitude.

```python
#load the data
years, magnitudes, longitude, latitude = np.load("earthquakes.npy").T
plt.subplots_adjust(hspace = 0.8, wspace = 0.4)
#plot the histgram of years
plt.subplot(221)
plt.hist(years, edgecolor= 'k')
plt.xlabel("years")
plt.ylabel("#incidents")
plt.title("#earthquakes each year")
#plot the histgram of magnitudes
plt.subplot(222)
plt.hist(magnitudes,bins = 50)
plt.xlabel("magnitudes")
plt.ylabel("#incidents")
plt.title("#earthquakes each magnitude")
#plot the scatterplot of location of all earthquakes
plt.subplot(223)
plt.scatter(longitude, latitude,alpha=.5, s = .1)
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.title("locations of earthquakes")
plt.axis("equal")
#plot the scatterplot of location of all earthquakes that are greater than 6
plt.subplot(224)
mask = magnitudes > 6
plt.scatter(longitude[mask], latitude[mask],alpha=.5, s = .1, color = 'r')
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.title("locations of earthquakes > 6")
plt.axis("equal")
plt.show()


```

### Problem 5

The _Rosenbrock function_ is defined as follows. $$f(x,y)=(1−x)^2 +100(y−x^2)^2$$
The minimum value of $f$ is $0$, which occurs at the point $(1, 1)$ at the bottom of a steep, banana-shaped valley of the function.

Use a heat map and a contour plot to visualize the Rosenbrock function.
Plot the minimizer $(1, 1)$.
Use a different sequential colormap for each visualization.

```python
#initialize the function and domain
x = np.linspace(-2,2, 200)
X, Y = np.meshgrid(x, x)
Z = (1-X)**2 +100*(Y-X**2)**2
plt.subplots_adjust(hspace = 0.8, wspace = 0.4)
#plot the Heat map
plt.subplot(221)
plt.plot(1,1, "o",ms=7)
plt.pcolormesh(X, Y, Z, cmap="viridis", norm=colors.LogNorm(vmin=1e-6))
plt.colorbar()
plt.xlabel("x")
plt.ylabel("y")
#plot the Contour map
plt.subplot(222)
plt.plot(1,1, "o",ms=5)
plt.contour(X, Y, Z, 6, cmap="viridis", norm=colors.LogNorm(vmin=1e-6))
plt.colorbar()
plt.xlabel("x")
plt.ylabel("y")
#plot the filledContour map
plt.subplot(223)
plt.plot(1,1, "o",ms=5)
plt.contourf(X, Y, Z, 6, cmap="viridis", norm=colors.LogNorm(vmin=1e-6))
plt.colorbar()
plt.xlabel("x")
plt.ylabel("y")

plt.show()

```

### Problem 6

The file `countries.npy` contains information from 20 different countries.
Each row in the array represents a different country; the columns are the 2015 population (in millions of people), the 2015 GDP (in billions of US dollars), the average male height (in centimeters), and the average female height (in centimeters), in that order.

Visualize this data set with at least four plots, using at least one scatter plot, one histogram, and one bar chart.
List the major insights that your visualizations reveal.

(Hint: consider using `np.argsort()` and fancy indexing to sort the data for the bar chart.)

```python
#initilize the data and the labels
countries = np.array(
            ["Austria", "Bolivia", "Brazil", "China", "Finland",
             "Germany", "Hungary", "India", "Japan", "North Korea",
             "Montenegro", "Norway", "Peru", "South Korea", "Sri Lanka",
             "Switzerland", "Turkey", "United Kingdom", "United States", "Vietnam"])
population, gdp, male, female = np.load("countries.npy").T
positions = np.arange(len(countries))



```

```python
plt.subplots_adjust(hspace = 0.8, wspace = 0.4)
#plot the correlation between the male femal heights.
plt.scatter(male, female, s = 3)
plt.xlabel("male(cm)")
plt.ylabel("female(cm)")
plt.title("correlation between femal and male heights")
domain = np.linspace(155,190,100)
s, i, r = lrg(male,female)[0:3]
plt.plot(domain, s*domain + i, "g")
plt.show()
#plot the GDP of each country(assending order)
index = np.argsort(gdp)
plt.barh(positions, gdp[index], align="center") # Horizontal bar char (better).
plt.yticks(positions, countries[index])
plt.tight_layout()
plt.xlabel("GDP(billion)")
plt.title("GDP of each country")
plt.show()
#plot the population of each country(assending order)
index = np.argsort(population)
plt.barh(positions, population[index], align="center") # Horizontal bar char (better).
plt.yticks(positions, countries[index])
plt.tight_layout()
plt.xlabel("population(million)")
plt.title("population of each country")
plt.show()
#plot the GDP per Capital of each country(assending order)
index = np.argsort(gdp/population)
plt.barh(positions, (gdp/population)[index], align="center") # Horizontal bar char (better).
plt.yticks(positions, countries[index])
plt.tight_layout()
plt.xlabel("GDP per Capital(billion/million)")
plt.title("GDP per Capital")
plt.show()
#plot the histgram of male and female
plt.hist(male, alpha=.5,bins = 20, label="Male")
plt.hist(female, alpha=.5, bins = 20, label="Female")
plt.xlabel("male/femal heights")
plt.title("male/femal heights")
plt.legend()
plt.show()
```
